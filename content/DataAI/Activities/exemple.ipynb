### Create a Series object from a list or NumPy array

```python
# From a list
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
```

```python
# From a NumPy array
import numpy as np
```

```python
data = np.array([10, 20, 30, 40, 50])
series = pd.Series(data)
```

```python
series = pd.Series([10, 20, 30, 40, 50])
```

```python
# Indexing
print(series[0])   # Access element at index 0
print(series[1:4]) # Slice elements from index 1 to 3
```

```python
# Mathematical operations
print(series.sum())     # Sum of all elements
print(series.mean())    # Mean of all elements
print(series.max())     # Maximum value in the series
print(series.std())     # Standard deviation of elements
```

## Dataframe

```python
data = {'Name': ['John', 'Alice', 'Bob', 'Alain'],
        'Age': [25, 28, 30, None],
        'Country': ['USA', 'Canada', 'UK', 'Canada']}

df = pd.DataFrame(data)
```

```python
# Selecting columns
print(df['Name'])       # Access the 'Name' column
print(df[['Name', 'Age']])  # Access multiple columns
```

```python
# Filtering rows
print(df[df['Age'] > 25])  # Filter rows based on a condition
```

```python
# Handling missing data
df.dropna()   # Drop rows with missing values
df.fillna(0)  # Fill missing values with 0
```

```python
# Create two example DataFrames
data1 = {'ID': [1, 2, 3],
         'Name': ['John', 'Alice', 'Bob']}
df1 = pd.DataFrame(data1)

data2 = {'ID': [2, 3, 4],
         'Age': [25, 30, 35]}
df2 = pd.DataFrame(data2)
```

```python
# Merge the DataFrames based on a common column 'ID'
merged_df = pd.merge(df1, df2, on='ID')
print(merged_df)
```

#### Grouping and aggregation operations on DataFrames

```python
# Create an example DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'John', 'Alice'],
        'City': ['New York', 'Paris', 'London', 'New York', 'Paris'],
        'Age': [25, 30, 35, 27, 32]}
df = pd.DataFrame(data)
```

```python
# Group the DataFrame by the 'Name' column and calculate the average age for each group
grouped_df = df.groupby(['Name', 'City']).mean()
print(grouped_df)
```

#### Descriptive statistics

```python
# Create an example DataFrame
data = {'Name': ['John', 'Alice', 'Bob'],
        'Age': [25, 30, 35],
        'Height': [175, 162, 180]}
df = pd.DataFrame(data)

# Compute descriptive statistics
print(df.describe())
```

## Data visualization using Pandas

```python
# We need pyplot from matplotlib, which is usually imported as plt
import matplotlib.pyplot as plt
```

```python
# Create an example DataFrame
data = {'Name': ['John', 'Alice', 'Bob'],
        'Age': [25, 30, 35],
        'Height': [175, 162, 180]}
df = pd.DataFrame(data)
```

```python
# Create a bar plot of age
df.plot(x='Name', y='Age', kind='bar')
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Distribution')
plt.show()
```
